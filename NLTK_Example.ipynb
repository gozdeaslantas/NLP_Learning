```
import nltk

nltk.corpus.gutenberg.fileids()
from nltk.corpus import gutenberg
alice_sentences = gutenberg.sents('carroll-alice.txt')
alice_sentences[2:6]
alice_raw = gutenberg.raw('carroll-alice.txt')
alice_raw[:100]

alice_words_raw = gutenberg.words('carroll-alice.txt')
fdist_raw = nltk.FreqDist(w.lower() for w in alice_words_raw)
fdist_raw.most_common(5)

#perform tokenization
nltk.download('punkt')
from nltk.tokenize import word_tokenize, sent_tokenize
print(sent_tokenize(alice_raw))
print(word_tokenize(alice_raw))

#remove punctuation
from string import punctuation
translator = str.maketrans('', '', punctuation)
pruned_sentence = alice_raw.translate(translator)

print(pruned_sentence[0:1000])
tokens = [word.lower() for sentence in nltk.sent_tokenize(alice_raw[:100]) for word in word_tokenize(pruned_sentence[0:1000])]

#tokens[0:10]

#perform stemming
porter = nltk.PorterStemmer()
[porter.stem(t) for t in tokens]


from wordcloud import WordCloud
import matplotlib.pyplot as plt

wordCloud = WordCloud().generate(' '.join(tokens))

plt.imshow(wordCloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
